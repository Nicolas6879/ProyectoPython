```python
import pygame
import random
import sys

# Inicialización de Pygame
pygame.init()

# Constantes del juego
SCREEN_WIDTH, SCREEN_HEIGHT = 800, 600
ENEMY_SIZE = 50
PLAYER_SIZE = 50
PLAYER_COLOR = (0, 255, 0)
ENEMY_COLOR = (255, 0, 0)
BACKGROUND_COLOR = (0, 0, 0)
SPEED = 10
MAX_ENEMIES = 15 

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
clock = pygame.time.Clock()  # Reloj para controlar la velocidad de actualización

# Funciones del juego
def draw_player(player_pos):
    pygame.draw.rect(screen, PLAYER_COLOR, (player_pos[0], player_pos[1], PLAYER_SIZE, PLAYER_SIZE))

def draw_enemy(enemy_pos):
    pygame.draw.rect(screen, ENEMY_COLOR, (enemy_pos[0], enemy_pos[1], ENEMY_SIZE, ENEMY_SIZE))

def detect_collision(player_pos, enemy_pos):
    p_x, p_y = player_pos
    e_x, e_y = enemy_pos
    margin = 0 #ajustar la dificultad (0=muy dificil, 50=no se puede perder)
    return (e_x + ENEMY_SIZE > p_x + margin and e_x < p_x + PLAYER_SIZE - margin) and (e_y + ENEMY_SIZE > p_y + margin and e_y < p_y + PLAYER_SIZE - margin)

def draw_text(text, font, color, x, y):
    text_surface = font.render(text, True, color)
    text_rect = text_surface.get_rect()
    text_rect.center = (x, y)
    screen.blit(text_surface, text_rect)
    
def calculate_enemy_count(elapsed_seconds):
    # Ajusta esta función según sea necesario para aumentar la cantidad de enemigos con el tiempo
    if elapsed_seconds < 10:
        return 5
    elif elapsed_seconds < 20:
        return 7
    elif elapsed_seconds < 30:
        return 10
    else:
        return MAX_ENEMIES

# Juego Principal
def game_loop():
    player_pos = [SCREEN_WIDTH//2, SCREEN_HEIGHT-2*PLAYER_SIZE]

    enemy_list = []
    running = False
    game_over = False
    pausado = False
    start_time = pygame.time.get_ticks()

    while not running:
        screen.fill(BACKGROUND_COLOR)
        draw_text("Space to Start", pygame.font.Font(None, 36), (255, 255, 255), SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2)
        pygame.display.update()
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:  # Cambiar el evento a la tecla de espacio
                    running = True

    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_ESCAPE:
                    running = False
                elif event.key == pygame.K_p:
                    pausado = not pausado
                elif event.key == pygame.K_SPACE and game_over:
                    # Reiniciar el juego
                    game_loop()
        
        
        if not pausado:            
            # Calcular el tiempo transcurrido
            elapsed_time = pygame.time.get_ticks() - start_time  # Tiempo en milisegundos
            elapsed_seconds = elapsed_time // 1000  # Convertir a segundos

            # Calcular la puntuación
            score = elapsed_seconds  # Puntuación igual al tiempo jugado en segundos
            
            # Calcular la cantidad de enemigos a generar
            enemy_count = calculate_enemy_count(elapsed_seconds)
            
            # Movimiento del jugador
            keys = pygame.key.get_pressed()
            if keys[pygame.K_LEFT] and player_pos[0] > 0:
                player_pos[0] -= SPEED
            if keys[pygame.K_RIGHT] and player_pos[0] < SCREEN_WIDTH - PLAYER_SIZE:
                player_pos[0] += SPEED

            # Generar enemigos
            if len(enemy_list) < MAX_ENEMIES:
                while len(enemy_list) < enemy_count:
                    x_pos = random.randint(0, SCREEN_WIDTH-ENEMY_SIZE)
                    y_pos = 0
                    enemy_list.append([x_pos, y_pos])

            # Actualizar posición de los enemigos
            for idx, enemy_pos in enumerate(enemy_list):
                if enemy_pos[1] >= 0 and enemy_pos[1] < SCREEN_HEIGHT:
                    enemy_pos[1] += SPEED
                else:
                    enemy_list.pop(idx)

            # Verificar colisión
            for enemy_pos in enemy_list:
                if detect_collision(player_pos, enemy_pos):
                    game_over = True
                    pausado=True
                    

        # Dibujar elementos en pantalla
        screen.fill(BACKGROUND_COLOR)
        draw_player(player_pos)
        for enemy_pos in enemy_list:
            draw_enemy(enemy_pos)
        draw_text("Score: {}".format(score), pygame.font.Font(None, 24), (255, 255, 255), 50, 50)
        
            
        if game_over==False:
            if pausado:
                font = pygame.font.SysFont(None, 36)
                text = font.render("Pausado", True, (255, 255, 255))
                screen.blit(text, (SCREEN_WIDTH//2 - text.get_width()//2, SCREEN_HEIGHT//2 - text.get_height()//2))
        else:
            draw_text("Game Over! Space  to Restart", pygame.font.Font(None, 36), (255, 255, 255), SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2)
            draw_text("Score: {}".format(score), pygame.font.Font(None, 24), (255, 255, 255), 50, 50)
        
        
        pygame.display.update()
        clock.tick(30)  # Limitar la velocidad de actualización a 30 FPS

if __name__ == "__main__":
    game_loop()
    pygame.quit()
    sys.exit()
```
